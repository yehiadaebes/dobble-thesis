```python
# list of card combinations for chosen number of objects
def proj(N):
    full_list = {i:[] for i in range(1,N**2+N+2)}
    for c in range(0,N+1):
        y = [N*c+b+2 for b in range(N)]
        y.insert(0,1)
        full_list[c+1] = y
    for c in range(N):
        for a in range(2,N+2):
            x = [a]
            dummy = list(range(N))
            for b in range(N):
                x.append(N*(b+1) + dummy[(c+(b)*(a-2))%N] + 2)
            full_list[N*(a-1)+2+c] = x
    return full_list
```

```python
# incidence matrix
import numpy as np
def inci(N):
    zero = [0]*(N**2+N+1)
    A = np.array([zero]*(N**2+N+1))
    for i in range(N**2+N+1):
        x = proj(N)[i+1]
        y = [0]*(N**2+N+1)
        for j in x:
            y[j-1] = 1
        A[i] = y
    return A
```

```python
# check if a certain order works
# if doesn't work, it gives pairs of cards (id values)
    # where there is more than one matching object
    # doesn't show how many matching objects, but does that matter?
# not entirely accurate? 8 and 9 didn't work? it's fine for all primes

from numpy import dot
found_problem = False

######
N = 7
######
inc = inci(N)
n = N**2+N+1
for i in range(n):
    for j in range(n):
        if dot(inc[i],inc[j]) != 1 and i != j:
            print('(i,j)=(' + str(i+1) + ',' + str(j+1) + ')')
            found_problem = True

if not found_problem:
    print('The game works. Every possible pair shares only one matching object.')
```
